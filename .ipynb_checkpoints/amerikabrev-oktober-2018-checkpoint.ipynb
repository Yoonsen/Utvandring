```python
def testord(words, exp=1.0):
    """Wrapper for the computation of letters to reference"""
    words = fix_caps(words)
    res = dict()
    df_combo = pd.DataFrame(combo['brev']**exp/combo['ref']).sort_values(by=0, ascending=False)
    for w in words:
        if w in df_combo.index:
            res[w] = df_combo.transpose()[w]
    return (pd.DataFrame.from_dict(res, orient='index').join(combo, how='outer')).sort_values(by=0, ascending=False)
```

```python
def word_collocations(words, urns=[], before=5, after=5, limit=1000):
    reiste = dict()
    for r in words:
        reiste[r] = nb.urn_coll(r, urns = abrev, before =10, after=10, limit=10000)
        reiste[r].columns =[r]
    frame = pd.DataFrame(pd.concat([reiste[r] for r in reiste], axis=1).fillna(0).sum(axis=1))
    tot = frame.sum()
    nb.normalize_corpus_dataframe(frame)
    return frame, tot
```

```python
def coll_graph(collword, frame, top=50):
    edges = [(collword, w, {'weight': frame[0][w]}) for w in frame[:top].index ]
    G = nx.Graph()
    G.add_edges_from(edges)
    return G
        
```

```python
def make_recursive_graph(word):
    temp, _ = word_collocations([word], urns=urns)
    rG = dict()
    rG[word] = coll_graph(word, sframe(temp[0]**1.5/combo['ref']))
    for n in set(rG[word].nodes()) - {word}:
        temp = word_collocations([n], urns=urns )
        rG[n] = coll_graph(n, sframe(temp[0]**1.5/combo['ref']))
    return rG
                
```

```python
def fix_caps(words):
    low_text = [x.lower() for x in words.split()]
    upp_text = [x.capitalize() for x in words.split()]
    return list(set(low_text + upp_text))
```
